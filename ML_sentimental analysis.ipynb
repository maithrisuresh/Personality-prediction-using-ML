```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ana = SentimentIntensityAnalyzer()

sentiment_score = []

for post in personality["cleaned_posts"]:
    score = ana.polarity_scores(post)
    sentiment_score.append(score)
```

```python
personality["compound--sentiment"] = [score["compound"] for score in sentiment_score]
personality["positive--sentiment"] = [score["pos"] for score in sentiment_score]
personality["negative--sentiment"] = [score["neg"] for score in sentiment_score]
personality["neutral--sentiment"] = [score["neu"] for score in sentiment_score]
```

```python
scaler = MinMaxScaler()
personality["compound--sentiment"] = scaler.fit_transform(
    np.array(personality["compound--sentiment"]).reshape(-1, 1))
personality["positive--sentiment"] = scaler.fit_transform(
    np.array(personality["positive--sentiment"]).reshape(-1, 1))
personality["negative--sentiment"] = scaler.fit_transform(
    np.array(personality["negative--sentiment"]).reshape(-1, 1))
personality["neutral--sentiment"] = scaler.fit_transform(
    np.array(personality["neutral--sentiment"]).reshape(-1, 1))
```

```python
def words(s):
    w = set(s.split(" "))
    return len(w) / 50
```

```python
def emojis(post):

    emcount = 0
    words = post.split()
    for e in words:
        if "http" not in e:
            if e.count(":") == 2:
                emcount += 1
    return emcount / 50
```

```python
def colons(post):
   
    cocount = 0
    words = post.split()
    for e in words:
        if "http" not in e:
            cocount += e.count(":")
    return cocount / 50
```

```python
personality["question_mark"] = personality["posts"].apply(lambda a: a.count("?") / 50)
personality["exclamation_marks"] = personality["posts"].apply(lambda b: b.count("!") / 50)
personality["colon"] = personality["posts"].apply(colons)
personality["emoji"] = personality["posts"].apply(emojis)
```

```python
personality["word"] = personality["posts"].apply(lambda c: (c.count(" ") + 1) / 50)
personality["unique_word"] = personality["posts"].apply(words)
```

```python
personality["links"] = personality["posts"].apply(lambda d: d.count("http") / 50)
```

```python
personality["images_count"] = [len(re.findall(r"(\.jpg)|(\.jpeg)|(\.gif)|(\.png)", post)) / 50 for post in personality["posts"]]
```
